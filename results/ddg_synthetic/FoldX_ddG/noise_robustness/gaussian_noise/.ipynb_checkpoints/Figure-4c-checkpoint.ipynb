```python
import numpy as np
import pandas as pd
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load data: model predictions and true values

```python
def calc_metrics(noise_scale):

    df = pd.read_csv(f"model_predictions/Graphinity_predictions_test-noise_robustness-gaussian_noise-{noise_scale}.csv")
            
    # pearson correlation
    pc = pearsonr(df["true_label"],df["pred_score"])[0]
    
    # distribution comparison - ratio of st devs
    sdr = df["pred_score"].std()/df["true_label"].std()
    if sdr > 1: sdr = df["true_label"].std()/df["pred_score"].std()

    return pc, sdr
```

```python
metrics = []
for noise_scale in [0,0.5,1,2,5,10]:
    pcs, sdr = calc_metrics(noise_scale)
    metrics.append([noise_scale, pcs, sdr])
```

```python
metrics_df = pd.DataFrame(metrics, columns=["Gaussian Noise Scale","Test Pearson's Correlation", "Standard Deviation Ratio"])
```

### Plot

##### Figure 4c middle

```python
fig,ax = plt.subplots(figsize=(6.75,5))

ax.plot(metrics_df["Gaussian Noise Scale"],metrics_df["Test Pearson's Correlation"],
        marker="x", markersize=8, markeredgewidth=1.5, color="#54278f", linestyle="-",
        label="Test Pearson's Correlation")
ax.plot(metrics_df["Gaussian Noise Scale"],metrics_df["Standard Deviation Ratio"],
         marker="+", markersize=10, markeredgewidth=1.5, color="#54278f", linestyle="--",
         label="Test Standard Deviation Ratio")

ax.set_ylabel("Test Metric", fontsize=12)
ax.set_xlabel("Scale of Added Gaussian Noise", fontsize=12)
ax.set_ylim(-0.1,1.1)
ax.tick_params(axis='both', which='major', labelsize=12)
ax.legend(frameon=False, bbox_to_anchor=(0.64,0.2), fontsize=12)

plt.title("Dataset: Synthetic_FoldX_ΔΔG_942723_gaussian_noise")

plt.savefig("Figure-4c-middle.png", dpi=300)
```

#### Figure 4c - left and right

```python
def load_and_plot(left_right):
    
    if left_right == "left": noise_scale = 1
    elif left_right == "right": noise_scale = 10
    
    df = pd.read_csv(f"model_predictions/Graphinity_predictions_test-noise_robustness-gaussian_noise-{noise_scale}.csv")
    
    # plot upper – scatter

    plt.scatter(df["true_label"],df["pred_score"],marker="x",color="k") # label = true, score = pred
    plt.xlabel(r"True FoldX $\Delta\Delta$G (kcal/mol)")
    plt.ylabel(r"Predicted $\Delta\Delta$G (kcal/mol)")
    # y=x line
    lims = [
        np.min([plt.xlim(), plt.ylim()]),  # min of both axes
        np.max([plt.xlim(), plt.ylim()]),  # max of both axes
    ]
    plt.plot(lims, lims, "k--", label="$y = x$")
    # trendline
    z = np.polyfit(df["true_label"],df["pred_score"],1)
    p = np.poly1d(z)
    plt.plot(df["true_label"],p(df["true_label"]), linestyle="-", color="#54278f", label="Trendline", linewidth=2)
    # Pearson correlation on plot
    if left_right == "left":
        plt.text(-68.5, 5,f"r = {pearsonr(df['true_label'],df['pred_score'])[0]:.2f}", ha="center", va="center")
    elif left_right == "right":
        plt.text(-69, 2,f"r = {pearsonr(df['true_label'],df['pred_score'])[0]:.2f}", ha="center", va="center")
    plt.tight_layout()
    plt.legend()
    plt.axis("square")

    ax = plt.gca()

    ax.set_xticks(np.arange(-60,40,20))
    ax.set_yticks(np.arange(-60,40,20))
    
    plt.savefig(f"Figure-4c-upper_{left_right}", dpi=300)
    plt.show()
    plt.clf()

    # plot lower – histogram
    
    sns.histplot(df["true_label"], bins=np.arange(-90,25,0.25), kde=True, alpha=0.25,
                 kde_kws={"bw_adjust":1.5}, color="k", label="True")
    sns.histplot(df["pred_score"], bins=np.arange(-90,25,0.25), kde=True, alpha=0.25,
                 kde_kws={"bw_adjust":1.5}, color="#54278f", label="Predicted")

    if left_right == "left":
        plt.text(x=-7.75, y=17500, s=f"Standard Deviation Ratio = {round(df['pred_score'].std()/df['true_label'].std(), 2)}")
    elif left_right == "right":
        plt.text(x=-7.75, y=29000, s=f"Standard Deviation Ratio = {round(df['pred_score'].std()/df['true_label'].std(), 2)}")

    plt.xlim(-8,5)
    plt.xlabel(r"FoldX $\Delta\Delta$G (kcal/mol)")
    plt.ylabel("Count")
    plt.legend(frameon=False)

    plt.tight_layout()

    plt.savefig(f"Figure-4c-lower_{left_right}", dpi=300)
    plt.show()
```

```python
load_and_plot("left")
```

```python
load_and_plot("right")
```
